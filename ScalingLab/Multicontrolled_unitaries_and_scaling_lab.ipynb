# **Introduction**

The number of elementary gates required to decompose a general unitary of n qubits on a quantum computer depends on the specific unitary being decomposed and the decomposition method used.

One popular method for decomposing a unitary gate into elementary gates is the Solovay-Kitaev algorithm, which was introduced in a paper by Daniel Gottesman and Isaac L. Chuang in 1999 [1]. This algorithm is based on the idea that any unitary gate can be approximated to arbitrary precision using a sequence of a small set of elementary gates. The algorithm uses a recursive approach to construct a sequence of gates that approximates the target unitary to a desired accuracy.

The number of elementary gates required by the Solovay-Kitaev algorithm depends on the desired accuracy of the approximation and the set of elementary gates used. In general, the number of gates required grows exponentially with the number of qubits. However, for many practical applications, it is possible to achieve accurate approximations using a relatively small number of gates.

Another method for decomposing a unitary gate into elementary gates is the Trotter-Suzuki decomposition, which was introduced by M. Suzuki in 1976 [2]. This method approximates the target unitary by breaking it down into a sequence of small time steps, each of which can be approximated using a small number of elementary gates. The accuracy of the approximation depends on the size of the time steps and the number of steps used.

The number of elementary gates required by the Trotter-Suzuki decomposition depends on the size of the time steps and the accuracy desired. In general, the number of gates required grows polynomially with the number of qubits.

In summary, the number of elementary gates required to decompose a general unitary gate on a quantum computer depends on the specific unitary being decomposed, the decomposition method used, and the desired accuracy of the approximation. The Solovay-Kitaev algorithm and the Trotter-Suzuki decomposition are two commonly used methods for decomposing unitary gates into elementary gates.

**Exercise 2:** do the same as before but using an arbitrary number of control qubits *(you will have to decompose multiple times before reaching elementary gates)*.

```python
n = 5
qc2 = QuantumCircuit(n)
qc2.mcx(control_qubits = list(range(n - 1)), 
        target_qubit = (n-1))
qc2.draw('mpl')
```

```python
lastdepth = qc2.depth()
d_qc2 = qc2.decompose()
steps = 1

while d_qc2.depth() > lastdepth:
    lastdepth = d_qc2.depth()
    steps += 1
    d_qc2 = d_qc2.decompose()
    
print(f"Iteration needed = {steps}")    
d_qc2.draw('mpl')
```

**Exercise 3:** repeat what done above for circuits of 5 up to 15 qubits (i.e 4 up to 14 controls and 1 target). Plot the depth of the circuit as a function of the number of qubits using mathplotlib. Show the depth scales eponentially with the number of qubits.

```python
gates_number = []
for n in range(3,16):
    print(f"\n Starting with {n} qubits")
    qc3 = QuantumCircuit(n)
    qc3.mcx(control_qubits = list(range(n - 1)), 
            target_qubit = (n-1))
    lastdepth = qc3.depth()
    d_qc3 = qc3.decompose()
    step = 1
    while d_qc3.depth() > lastdepth:
        lastdepth = d_qc3.depth()
        step += 1 
        d_qc3 = d_qc3.decompose()
        print(f"Depth at step {step} = {d_qc3.depth()}")
    gates_number.append(d_qc3.depth())
    print(f"Iteration needed for {n} qubits: {step}, minimum gates needed {d_qc3.depth()}")  
```

```python
sns.set_palette('Set2')
def func(x, a, b, c):
    return a * np.exp(b * x) + c 
    
x = np.array((list(range(3,16))))
y = gates_number

#yn = y + 0.2 * np.random.normal(size=len(x))
plt.scatter(list(range(3,16)), gates_number, color=sns.color_palette("Set2")[1])
popt, pcov = curve_fit(func, x, y)
a, b, c = popt
plt.plot(x, func(x, a, b, c))
```
